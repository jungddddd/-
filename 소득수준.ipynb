```python
import pandas as pd
import numpy as np
import os
```

```python
fol = "B058. 서울시 행정동단위 분기별 추정매출액/2. 파일데이터"
data_list = []

for subfol in os.listdir(fol):
    subfol_path = os.path.join(fol, subfol)
    
    if os.path.isdir(subfol_path):
        for file in os.listdir(subfol_path):
            file_path = os.path.join(subfol_path, file)
            df = pd.read_csv(file_path, encoding='utf-8-sig', sep='|')
            data_list.append(df)
            
income = pd.concat(data_list, ignore_index=True)
income
```
